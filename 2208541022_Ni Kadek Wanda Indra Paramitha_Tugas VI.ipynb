## 5.21.2 Properti pada Pandas

```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#BEBERAPA PROPERTI PADA PANDAS
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

print(bunga_iris.head(4))
print('\n')
#Fungsi "head()" digunakan untuk menampilkan sejumlah baris awal dari dataframe. 
#Jika kita tidak memberikan parameter pada fungsi "head()", maka outputnya akan menampilkan 5 baris pertama dari dataframe. 
#Namun, dalam kode di atas, kita memberikan parameter "4" pada fungsi "head()", 
#sehingga hasil output yang dihasilkan adalah 4 baris pertama dari dataframe.

print(bunga_iris.columns)
print('\n')
#Fungsi "columns" digunakan untuk menampilkan nama-nama kolom dari dataframe. 
#Dalam kode di atas, saat kita memanggil fungsi "columns" pada dataframe "bunga_iris", 
#maka output yang dihasilkan adalah daftar nama kolom pada dataframe tersebut.

print(bunga_iris['sepal length (cm)'])
print('\n')
#kita dapat memilih kolom tertentu dari dataframe,
#dengan cara menggunakan tanda kurung siku "[]" diikuti dengan nama kolom yang ingin dipilih. 
#Pada kode di atas, kita memilih kolom "sepal length (cm)" dari dataframe "bunga_iris",
#dengan cara menuliskan nama kolom tersebut di dalam tanda kurung siku setelah nama dataframe.
#maka hasil output yang dihasilkan adalah data pada kolom "sepal length (cm)" dari dataframe "bunga_iris". 

print(bunga_iris['sepal length (cm)'][0])
print('\n')
#kita menambahkan index dengan menuliskan angka 0 di dalam tanda kurung siku setelah nama kolom. 
#Angka 0 ini menunjukkan bahwa kita ingin memilih nilai pada baris pertama dari kolom "sepal length (cm)".
#Hasil output yang dihasilkan adalah nilai pada baris pertama dan kolom "sepal length (cm)" dari dataframe "bunga_iris".

print(bunga_iris['sepal length (cm)'][1:5])
print('\n')
#Range angka 1:5 ini menunjukkan bahwa kita ingin memilih nilai pada baris kedua hingga kelima dari kolom "sepal length (cm)".
#output yang dihasilkan adalah nilai pada baris kedua hingga kelima dan kolom "sepal length (cm)" dari dataframe "bunga_iris".

print(bunga_iris[['sepal length (cm)', 'sepal width (cm)']][1:5])
print('\n')
#Kode di atas digunakan untuk memilih dua kolom 'sepal length (cm)' dan 'sepal width (cm)' 
#dari baris kedua hingga kelima dari dataframe "bunga_iris" menggunakan indexing pada Pandas.

print(bunga_iris.iloc[1])
print('\n')
#Angka 1 pada kode tersebut menunjukkan bahwa kita ingin memilih baris dengan index ke-1 pada dataframe. 
#Perlu diingat bahwa di Python, indeks dimulai dari angka 0.
#maka output yang dihasilkan adalah data pada baris kedua dari dataframe "bunga_iris". 

print(bunga_iris.iloc[1:5])
print('\n')
#Range angka 1:5 pada kode tersebut menunjukkan bahwa kita ingin memilih baris dengan index ke-1 hingga ke-4 pada dataframe. 
#maka output yang dihasilkan adalah data pada baris kedua hingga kelima dari dataframe "bunga_iris". 

print(bunga_iris.loc[bunga_iris['sepal length (cm)'] >= 6.0])
print('\n')
#Pada kode di atas, kita menggunakan method "loc" pada dataframe "bunga_iris",
#untuk memilih baris yang memiliki nilai sepal length lebih besar atau sama dengan 6.0. 
#Kita menggunakan operator pembanding ">=" untuk membandingkan setiap nilai pada kolom "sepal length (cm)" dengan nilai 6.0.
#sehingga output yang dihasilkan adalah data pada baris-baris yang memiliki nilai sepal length lebih besar atau sama dengan 6.0.

#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#output :
```

## 5.21.3 Mengurutkan data pada Pandas

```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#Data yang telah dipanggil dapat diurutkan mulai dari tinggi ke rendah (atau dari rendah ke tinggi) untuk data berupa angka, 
#sedangkan data berupa string akan diurutkan berdasarkan urutan abjad.

print(bunga_iris.sort_values('sepal length (cm)'))
print('\n')
#Pada kode di atas, kita menggunakan metode "sort_values()" pada dataframe "bunga_iris" 
#untuk mengurutkan nilai dalam dataframe berdasarkan kolom "sepal length (cm)". 
#output yang dihasilkan adalah dataframe "bunga_iris" yang telah diurutkan berdasarkan nilai pada kolom "sepal length (cm)" 
#dari nilai yang terkecil hingga nilai yang terbesar.

print(bunga_iris.sort_values('sepal width (cm)', ascending= False))
print('\n')
#Kita menambahkan parameter "ascending=False" pada metode "sort_values()" untuk mengurutkan nilai dalam urutan yang terbalik.
#output yang dihasilkan adalah dataframe "bunga_iris" yang telah diurutkan berdasarkan nilai pada kolom "sepal width (cm)" 
#dari nilai yang terbesar hingga nilai yang terkecil.

print(bunga_iris.sort_values(['petal length (cm)', 'petal width (cm)']))
print('\n')
#Pada kode di atas, kita menggunakan metode untuk mengurutkan nilai dalam dataframe berdasarkan beberapa kolom, 
#yaitu "petal length (cm)" dan "petal width (cm)".
#output yang dihasilkan adalah dataframe "bunga_iris" yang telah diurutkan berdasarkan nilai pada kolom "petal length (cm)" 
#dari nilai yang terkecil hingga nilai yang terbesar, dan jika terdapat nilai yang sama pada kolom "petal length (cm)", 
#maka akan diurutkan berdasarkan nilai pada kolom "petal width (cm)" dari nilai yang terkecil hingga nilai yang terbesar.

#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#output:
```

## 5.21.4 Mengubah data pada Pandas

```python
bunga_iris['5x Sepal Length (cm)'] = 5*bunga_iris['sepal length (cm)']
bunga_iris

#Kode di atas digunakan untuk menambahkan sebuah kolom baru dengan nama "5x Sepal Length (cm)" pada dataframe "bunga_iris" 
#dengan nilai yang dihasilkan dari operasi perkalian antara nilai pada kolom "sepal length (cm)" dengan angka 5.
```

```python
bunga_iris = bunga_iris.drop(columns = ['5x Sepal Length (cm)'] )
bunga_iris
#Pada kode di atas, kita menggunakan method "drop" pada dataframe "bunga_iris" dengan memberikan parameter "columns" 
#yang berisi nama kolom yang ingin dihapus yaitu "5x Sepal Length (cm)". 
#Setelah kode di atas dijalankan, dataframe "bunga_iris" akan tidak memiliki kolom "5x Sepal Length (cm)" lagi.
```

```python
bunga_iris['Luas Sepal'] = bunga_iris['sepal length (cm)']*bunga_iris['sepal width (cm)']
bunga_iris

#Pada kode di atas, kita menggunakan tanda kurung siku [ ] pada dataframe "bunga_iris" untuk memanggil kolom "sepal length (cm)"
#dan "sepal width (cm)" lalu mengalikan nilai-nilai dari kedua kolom tersebut dengan menggunakan operator perkalian (*). 
#Kemudian, kita menambahkan kolom baru dengan nama "Luas Sepal" menggunakan tanda sama dengan (=) 
#dan memasukkan hasil operasi tersebut pada kolom baru tersebut.
#Setelah kode di atas dijalankan, dataframe "bunga_iris" akan memiliki kolom baru bernama "Luas Sepal" 
#yang berisi nilai-nilai yang dihasilkan dari operasi perkalian antara nilai pada kolom "sepal length (cm)" 
#dengan nilai pada kolom "sepal width (cm)".
```

```python
bunga_iris['Luas Sepal'] = bunga_iris.iloc[:,3:7].sum(axis=1)
bunga_iris

#Kode di atas digunakan untuk menambahkan sebuah kolom baru dengan nama "Luas Sepal" pada dataframe "bunga_iris" 
#dengan nilai yang dihasilkan dari operasi penjumlahan antara nilai pada kolom ke-3 hingga kolom ke-6.
```

## 5.21.6 Menyaring data menggunakan Pandas

```python
bunga_iris.loc[bunga_iris['sepal length (cm)'] == 5.0]
#Kode di atas akan menghasilkan subset dari dataframe "bunga_iris" 
#yang hanya terdiri dari baris yang memiliki nilai "sepal length (cm)" sama dengan 5.0. 
#Jika tidak ada baris yang memenuhi kondisi tersebut, maka subset yang dihasilkan akan kosong.
```

```python
bunga_iris.loc[(bunga_iris['sepal length (cm)'] == 5.0) & 
               (bunga_iris['petal length (cm)'] == 1.6)]

#Kode tersebut digunakan untuk melakukan seleksi data pada kolom 'sepal length (cm)' dengan nilai 5.0 
#dan pada kolom 'petal length (cm)' dengan nilai 1.6 secara bersamaan. 
#Fungsi 'loc' pada Pandas digunakan untuk memilih baris dan kolom berdasarkan label atau kondisi tertentu. 
#Pada kode di atas, digunakan operator logika "AND" untuk memastikan kedua kondisi terpenuhi sebelum melakukan seleksi.
```
